# Bai 1: Tinh gia tri ham so

```python
# Khai bao thu vien
import math

# Nhap du lieu
x = float(input("Moi ban nhap bien so x: "))

# Xu ly
f_x = x + x ** 5 / math.factorial(5) + math.sqrt(abs(x)) / (x**(3/2))

# Xuat du lieu
print(f'Gia tri cua ham so f({x}) = {f_x:.2f}.')
```

# Bai 2: Doi sang gio phut giay

```python
# Khai bao thu vien
import math

# Nhap du lieu
t= int(input("Nhap vao tong so giay: "))

# Xu ly
hh = t // 3600
mm = (t % 3600) // 60
ss = t % 60

# Xuat du lieu
print(f'{t} giay co dang {hh}:{mm}:{ss}')
```
