## Bai 1: Tinh toan tren day so

### 1. Tao mot day so co N phan tu (N = 10)

```python
import numpy as np

print("1. Tao mot day so a co N phan tu (N = 10)")
a = np.array([120, 183, 160, 175, 145, 162, 190, 160, 152, 162])
print("Day so a: ", a)
```

### 2. Cac thao tac tren day so a
a. Binh phuong cac phan tu trong day so <br/>
b. Tim so luong cac phan tu cua day so <br/>
c. In ra gia tri lon nhat, gia tri nho nhat va gia tri trung binh <br/>
d. Tính phương sai của dãy số $\sigma^2 = \frac{\sum{\left(a_i - \mu\right)^2}}{N - 1}$ với $\mu = \frac{\sum{a_i}}{N}$ và độ lệch chuẩn $var = \sqrt{\sigma^2}$

```python
print("2. Cac thao tac tren day so")

a2 = a**2
print("a. Binh phuong a: ", a2)

na = len(a)
print("\nb. Do ai cua a: ", na)

v_max = np.max(a)
v_min = np.min(a)
v_avg = np.mean(a)
print(f'\nc. Gia tri lon nhat: [{v_max}], gia tri nho nhat: [{v_min}], gia tri trung binh: [{v_avg}]')

v_sigma2 = np.var(a, ddof=1)
v_var = np.sqrt(v_sigma2)
print(f'\nd. Phuong sai: [{v_sigma2: .2f}] va do lech chuan: [{v_var: .2f}]')
```
